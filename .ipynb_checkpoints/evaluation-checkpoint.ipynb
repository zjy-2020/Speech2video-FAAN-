```python
'''PSNR  SSIM'''
import os
import numpy as np
import math
from PIL import Image
import cv2
import skimage
from skimage.measure import compare_ssim 
import os
import numpy as np
import math
from PIL import Image

def psnr(img1, img2):
    
    img1=cv2.resize(img1, (200, 200))
    img2=cv2.resize(img2, (200, 200))
    mse = np.mean((img1  - img2 ) ** 2)
    
    if mse < 1.0e-10:
        return 100 * 1.0
    return 20* math.log10(255.0 / math.sqrt(mse))


def mse(img1, img2):
    img1=cv2.resize(img1, (200, 200))
    img2=cv2.resize(img2, (200, 200))
    mse = np.mean((img1 / 1. - img2 / 1.) ** 2)
    return mse

def ssim(y_true, y_pred):
    u_true = np.mean(y_true)
    u_pred = np.mean(y_pred)
    var_true = np.var(y_true)
    var_pred = np.var(y_pred)
    std_true = np.sqrt(var_true)
    std_pred = np.sqrt(var_pred)
    c1 = np.square(0.01 * 255)
    c2 = np.square(0.03 * 255)
    ssim = (2 * u_true * u_pred + c1) * (2 * std_pred * std_true + c2)
    denom = (u_true ** 2 + u_pred ** 2 + c1) * (var_pred + var_true + c2)
    return ssim / denom


path1 = '/.../'  
path2 = '/.../'  
f_nums = len(os.listdir(path1))
list_psnr = []
list_ssim = []
list_mse = []
for i in range(1, 30):

    img_a = Image.open(path1 + str(i) + '.jpg')
    img_b = Image.open(path2 + str(i) + '.jpg')
   
    img_a = np.array(img_a)
    
    img_b = np.array(img_b)

    psnr_num = psnr(img_a, img_b)
    img_a=cv2.resize(img_a, (200, 200))
    img_b=cv2.resize(img_b, (200, 200))
  
    ssim_num = skimage.metrics.structural_similarity(img_a, img_b, multichannel=True)
    mse_num = mse(img_a, img_b)
    list_ssim.append(ssim_num)
    list_psnr.append(psnr_num)
    list_mse.append(mse_num)
print("average PSNR:", sum(list_psnr) / len(list_psnr) )  
print("average SSIM:", sum(list_ssim) / len(list_ssim))  
print("average MSE:", np.mean(list_mse))  
```
